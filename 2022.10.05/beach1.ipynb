```python
import os
import pandas as pd
```

```python
all_path = os.listdir('./data11/')
```

```python
all_path
```

```python
df= pd.DataFrame({'일시': pd.date_range('2017-01-01', '2022-09-30', freq='D')})
df['일시'] = df['일시'].astype('str')
df.set_index('일시', inplace=True)

for file in all_path:
    all=pd.read_csv('./data11/'+file, encoding='cp949')
    all.set_index('일시', inplace=True)
    df=pd.concat([df, all], axis=1)
```
