```python
import numpy as np
import pandas as pd
from tensorflow.keras import Sequential
from sklearn.model_selection import train_test_split
from tensorflow.keras.datasets.fashion_mnist import load_data
from tensorflow.keras.layers import Dense, Dropout, Flatten
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.utils import all_estimators
import warnings
warnings.filterwarnings('ignore')
```

```python
# data, target 지정
data=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
        'BMI', 'DiabetesPedigreeFunction', 'Age']]
target=df['Outcome']
```

```python
# 훈련 테스트 분리
train_X, test_X, train_y, test_y = train_test_split(data, target,
                                                    test_size=0.2,
                                                   shuffle=True,
                                                    stratify=target,
                                                   random_state=42)
```

```python
# 딥러닝 모델 설정
model=Sequential()
```

```python
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 딥러닝 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(train_X, train_y, epochs=100, batch_size=10)
```

```python
# 저장된 값 확인
print(train_X.iloc[0])
print('-----------------------------------------')
print(train_y.iloc[0])
```

```python
all_Algorithm = all_estimators(type_filter = 'regressor')
scaler_list = [StandardScaler(),MinMaxScaler()]
best_r2=[]
for scaler in scaler_list:
    scaler.fit(train_X)
    train_X = scaler.transform(train_X)
    test_X = scaler.transform(test_X)
    for (name,algorithm) in all_Algorithm:
        try:
            model = algorithm()
            model.fit(train_X, train_y)
            y_pred = model.predict(test_X)
            r2 = r2_score(test_y, y_pred)
            best_r2.append(r2)
            print(name , "\'s R2 : ", r2)
            print("-----------------------------------------------")
            print(r2,max(best_r2))
            if max(best_r2)<=r2:
                final_r2 = r2
                final_r2_model = name
                final_r2_scaler = scaler
        except:
            continue
        
print("Best R2 & model & scaler: ",final_r2," & ",final_r2_model, "&",scaler)
```
