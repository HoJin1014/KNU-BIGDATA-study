```python
import pandas as pd
```

```python
# re(정규표현식) => \s : 화이트스페이스
df=pd.read_csv(URL, sep='\s+', header=None)
```

```python
df.columns = ['mpg','cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
              'model year', 'origin', 'car name']
df.head()
```

```python
# print(df['horsepower'].value_counts())
```

```python
#origin
print(df["origin"].value_counts())
print(df["origin"].value_counts(normalize=True))
```

```python
#model-year
print(df["model year"].value_counts())
print(df["model year"].value_counts(normalize=True))
```

```python
#cylinders
print(df["cylinders"].value_counts())
print(df["cylinders"].value_counts(normalize=True))
```

```python
#convert object to numeric
df.head()
```

```python
#convert object to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
print(df.dtypes)
```

```python
#fillna-with-median
df["horsepower"].fillna(df['horsepower'].median)
```
