```python
import pandas as pd
```

```python
DIR1='./l.csv'
DIR2='./m.csv'
DIR3='./n.csv'
DIR4='./o.csv'
DIR5='./p.csv'
DIR6='./q.csv'
```

```python
df1=pd.read_csv(DIR1, header=None)
df2=pd.read_csv(DIR2, header=None)
df3=pd.read_csv(DIR3, header=None)
df4=pd.read_csv(DIR4, header=None)
df5=pd.read_csv(DIR5, header=None)
df6=pd.read_csv(DIR6, header=None)
```

```python
df1.drop_duplicates(subset=0, keep='last', inplace=True)
df2.drop_duplicates(subset=0, keep='last', inplace=True)
df3.drop_duplicates(subset=0, keep='last', inplace=True)
df4.drop_duplicates(subset=0, keep='last', inplace=True)
df5.drop_duplicates(subset=0, keep='last', inplace=True)
df6.drop_duplicates(subset=0, keep='last', inplace=True)
```

```python
# 1. Log 파일 합치기
dfall=pd.concat([df1,df2,df3,df4,df5,df6], ignore_index= 'bool', join='inner', axis=1)
```
