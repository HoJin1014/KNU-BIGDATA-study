```python
# 모듈 로딩 -------------------------------------------------------
import matplotlib.pyplot as plt
```

```python
# Matplotlib 한글 Font 설정 ------------------------
from matplotlib import font_manager as fm, rc

# 한글 폰트 설정
font_path='../Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 그래프 그리기 => plot(y_data)

plt.plot([1,2,3,4],[11,22,33,44])
plt.ylabel(['TEST DATA'])
plt.title('TEST MATPLOT')
plt.show()
```

```python
# 그래프 그리기 => plot([x_data](y_data)

plt.plot([1,2,3,4],[11,22,33,44])
plt.ylabel('TEST DATA')                 # y축 라벨
plt.xlabel('VALUE')
plt.title('TEST MATPLOT')                 # 그래프 타이틀
plt.legend('a')                           # 범례
plt.show()
```

```python
# 그래프 그리기 => plot([x_data.](y_data)

plt.plot([1,2,3,4],[11,22,33,44], 'r--')
plt.plot([1,2,3,4],[11,22,33,44], 'g^', [1,2,3,4],[11,33,55,77], 'bo')
plt.ylabel('TEST DATA')                   # y축 라벨
plt.xlabel('VALUE')                       # x출 라벨
plt.title('TEST MATPLOT')                 # 그래프 타이틀
plt.legend('a')                           # 범례
plt.show()
```

```python
# 여러개 그래프를 하나에 넣기 => subplot()
fig=plt.figure()

ax1 = fig.add_subplot(221)
ax1.set_title('LINE')
ax1.plot([1,2,3,4])

ax2 = fig.add_subplot(222)
ax2.set_title('BAR')
ax2.bar([1,2,3,4], [11,22,33,44], color='y')

ax3 = fig.add_subplot(223)
ax3.set_title("LINE")
ax3.plot([1,2,3,4])

# 여러개 plot 간격 자동 조절
fig.tight_layout()

# 화면 출력
plt.show()
```
