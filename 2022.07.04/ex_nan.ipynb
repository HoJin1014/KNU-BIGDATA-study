# 결측치(Missing Value) 처리 실습 --------

- 결측치(Missing Value)란?
    : 값이 들어있지 않는 것으로 NaN, NA로 표기
    
- 관련 메서드
    - 체크 : isnull(), isna(), notnull(), notna()
    - 삭제 : dropna()
    - 치환 : fillna()

```python
# 모듈 로딩 ---------------------------------------------------
import pandas as pd

# 파일관련 변수 선언 ------------------------------------------
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'weather.csv'

FILE_NAME
```

```python
# (1) CSV FILE => DateFrame으로 로딩 ---------------------------
weatherDF=pd.read_csv(FILE_NAME)
weatherDF
```

```python
# (2) 데이터 정보 확인
weatherDF.info()
```

```python
weatherDF.head(5)
```

```python
weatherDF.isnull()
```

```python
# 결측치 갯수 파악 ---------------------------------------------
weatherDF.isnull().sum()
```

```python
# 2010년도 1월의 최고 온도 - 최저 온도 출력 ---------------------
# (1) 1월 데이터만 추출
oneMonth=weatherDF.iloc[[9,1]]
oneMonth
```

```python
# (2) NaN값 제거 => dropna() : 기본 값 => 행(row)방향, 1개라도 NaN있으면 삭제
oneMonth.dropna()
```

```python
# => 행(column)방향 설정, 최고/최저온도값이 모두 NaN인것 삭제하고 싶음
# => dropna(axis=1, how='all')
oneMonth.dropna(axis=1, how='all')
```

```python
# 2월의 최고-최저온도 출력
twoMonth=weatherDF.iloc[[2,3]]
twoMonth
```

```python
# 2월의 최고-최저온도 출력
twoMonth=weatherDF.iloc[[2,3]].dropna(axis=1, how='all')
twoMonth
```

# 처리방법- (2) 치환

```python
# (1) CSV FILE => DateFrame으로 로딩 ---------------------------
weatherDF=pd.read_csv(FILE_NAME)
weatherDF
```

```python
# (2) NaN 데이터를 다른 값으로 치환(변경)하기 => fillna()
# 모든 NaN을 0으로 채우기
weatherDF.fillna(0)
```

```python
weatherDF.fillna(method='bfill')
weatherDF
```
