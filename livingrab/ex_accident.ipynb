```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
```

```python
# 네이버 페이지 교통사고 크롤링
genre1 = []

for i in range(10):
    print(f'{i}페이지 분석중....')
    url1 = (f'https://search.naver.com/search.naver?display=15&f=&filetype=0&page={i+2}&query=%EA%B5%90%ED%86%B5%EC%82%AC%EA%B3%A0&research_url=&sm=tab_pge&start={i*15+1}&where=web')
    html1 = requests.get(url1)
    bs = BeautifulSoup(html1.text, 'lxml')
    bs2 = bs.find('div', {'class':'api_txt_lines'}).get_text()
    genre1.append(bs2)
```

```python
print(genre1)
```

```python
# 네이버 view
url2 = 'https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%EA%B5%90%ED%86%B5%EC%82%AC%EA%B3%A0'
html2 = requests.get(url2)
bs3 = BeautifulSoup(html2.text, 'lxml')
bs4 = bs3.find_all('div', {'class':'api_txt_lines'})
bs4
```

```python
bs4_list = []
for i in bs4:
    bs4_list.append(i.get_text())
```

```python
print(bs4_list)
```

```python
genre1.extend(bs4_list)
```

```python
from konlpy.tag import Okt

okt = Okt()
noun_adj_list = []
for data in genre1:
    sentences_tag = okt.pos(data)
    # tag가 명사인 단어들만 noun_adj_list에 넣어준다.
    for word, tag in sentences_tag:
        if tag in ['Noun']:
            noun_adj_list.append(word)
```

```python
for i in range(71):
    noun_adj_list.remove('교통사고')
# noun_adj_list.remove('교통사고')
# del noun_adj_list[0]
# noun_adj_list
```

```python
noun_adj_list
counts = Counter(noun_adj_list)
tags = counts.most_common(50)
tags
```

```python
# WordCloud를 생성한다.
from wordcloud import WordCloud
import platform
import matplotlib.pyplot as plt

if platform.system() == 'Windows':
    path = r'c:\Windows\Fonts\malgun.ttf'
elif platform.system() == 'Darwin': # Mac OS
    path = r'/System/Library/Fonts/AppleGothic'
else:
    path = r'/usr/share/fonts/truetype/name/NanumMyeongjo.ttf'

wc = WordCloud(font_path=path, background_color="white", max_font_size=60)
cloud1 = wc.generate_from_frequencies(dict(tags))

# 생성된 WordCloud를 test.jpg로 보낸다.
#cloud.to_file('test.jpg')

plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(cloud1)
plt.show()
```
