```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import numpy as np
```

```python
# 네이버 페이지 교통사고 크롤링
genre1 = []

for i in range(10):
    print(f'{i}페이지 분석중....')
    url1 = (f'https://search.naver.com/search.naver?display=15&f=&filetype=0&page={i+2}&query=%EA%B5%90%ED%86%B5%EC%82%AC%EA%B3%A0&research_url=&sm=tab_pge&start={i*15+1}&where=web')
    html1 = requests.get(url1)
    bs = BeautifulSoup(html1.text, 'lxml')
    bs2 = bs.find('div', {'class':'api_txt_lines'}).get_text()
    genre1.append(bs2)
```

```python
print(genre1)
```

```python
# 네이버 view
url2 = 'https://search.naver.com/search.naver?where=view&sm=tab_jum&query=%EA%B5%90%ED%86%B5%EC%82%AC%EA%B3%A0'
html2 = requests.get(url2)
bs3 = BeautifulSoup(html2.text, 'lxml')
bs4 = bs3.find_all('div', {'class':'api_txt_lines'})
bs4
```

```python
bs4_list = []
for i in bs4:
    bs4_list.append(i.get_text())
```

```python
print(bs4_list)
```

```python
genre1.extend(bs4_list)
```

```python
from konlpy.tag import Okt

okt = Okt()
noun_adj_list = []
for data in genre1:
    sentences_tag = okt.pos(data)
    # tag가 명사인 단어들만 noun_adj_list에 넣어준다.
    for word, tag in sentences_tag:
        if tag in ['Noun']:
            noun_adj_list.append(word)
```

```python
def fill_mean_year(data:pd.DataFrame):
    data_c = data.copy()
    columns = data_c.columns[2:]
    for col in columns:
        df_replce_comma = data_c[col].str.replace(',', '')
        df_replace_dash = df_replce_comma.replace("-", np.NAN).astype(float)
        fill_mean = df_replace_dash.mean()
        df_replace_dash.fillna(fill_mean, inplace=True)
        data_c[col] = df_replace_dash
        
    data_c["교통사고건수"] = data_c["교통사고건수"].astype(int)
        
    return data_c
```

```python
df2015 = fill_mean_year(df2015)
df2016 = fill_mean_year(df2016)
df2017 = fill_mean_year(df2017)
df2018 = fill_mean_year(df2018)
df2019 = fill_mean_year(df2019)
```
