```python
# BeautifulSoup  
from bs4 import BeautifulSoup
url = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjsz3ZBxPY'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
def scraping_use_find(html):
    
    soup = BeautifulSoup(html.text, 'html.parser')
    soup1 = soup.find('div', {'id' : "seven-day-forecast"})

    period_name = soup1.find_all('p', {'class': 'period-name'} )
    short_desc = soup1.find_all('p',{'class': 'short-desc'} )
    img = soup1.find_all('img' )
    temp = soup1.find_all('p', {'class': 'temp'} )

    print('[find ]')
    for i in range(9):
        image = img[i]['title']
        print('------------------------------------------------------------------------------')
        print(f'[Period]: {period_name[i].text}') 
        print(f'[Short desc]: {short_desc[i].text}') 
        print(f'[Temperature]: {temp[i].text}') 
        print(f'[Image desc]: {image}') 

scraping_use_find(html)
```
